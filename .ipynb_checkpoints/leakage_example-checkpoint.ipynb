```python
import pandas as pd
from glob import glob
```

```python
# KLUE 데이터 출처입니다.
# https://aistages-prod-server-public.s3.amazonaws.com/app/Competitions/000068/data/klue-nli-v1.1.tar.gz
# 총 3개의 json파일이 있습니다. 제공된 파일 모두를 불러와 확인했습니다.
```

```python
original = pd.read_csv("./RawData/train_data.csv").drop("index", axis = "columns")
test_data = pd.read_csv("./RawData/test_data.csv").drop('index', axis = "columns")

# maybe lead to data leakage
external_jsons = glob("./RawData/external_data/*.json")
external_klue = []
for f in external_jsons :
    temp = pd.read_json(f).loc[:, ["premise", "hypothesis", "gold_label"]]
    external_klue.append(temp)
external_klue = pd.concat(external_klue)
external_klue = external_klue.rename(columns = {"gold_label" : "label"})
```

```python
original_text = original.premise + ' ' + original.hypothesis
test_text = test_data.premise + ' ' + test_data.hypothesis
external_text = external_klue.premise + ' ' + external_klue.hypothesis
```

```python
# Dacon측에서 제공한 학습용 데이터와 KLUE dataset 간 겹치는 데이터입니다. (출처가 KLUE이니 이는 문제 없을 듯 합니다.)
set(original_text) & set(external_text)
```

```python
# Dacon측에서 제공한 리더보드 제출용 데이터와 KLUE dataset 간 겹치는 데이터입니다.
# premise와 hypothesis 모두 겹치는 데이터가 총 6건이 존재합니다.
# KLUE에서 제공하는 데이터에 Label까지 존재함을 감안한다면 충분히 Leakage가 발생할 수 있다고 생각합니다.

set(test_text) & set(external_text)
```
